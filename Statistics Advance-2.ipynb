<h3 style='color: blue;'>Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.</h3>
<p><strong>Answer:</strong></p>
 <h3>Probability Mass Function (PMF) and Probability Density Function (PDF)</h3>

<h4>Probability Mass Function (PMF)</h4>
    <p>The Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value. The PMF is applicable only to discrete random variables.</p>
    <p>Formally, for a discrete random variable \(X\), the PMF \(P(X = x)\) is defined as:</p>
    <pre>P(X = x) = P(X \text{ takes the value } x)</pre>

 <h3>Example</h3>
    <p>Consider a fair six-sided die. Let \(X\) be the random variable representing the outcome of a single roll. The PMF of \(X\) is:</p>
    <pre>
    P(X = x) = 
    \begin{cases} 
    \frac{1}{6} & \text{if } x \in \{1, 2, 3, 4, 5, 6\} \\
    0 & \text{otherwise} 
    \end{cases}
    </pre>
    <p>This means that the probability of rolling any one of the six faces is \(\frac{1}{6}\), and the probability of rolling any other value is 0.</p>

 <h4>Probability Density Function (PDF)</h4>
    <p>The Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable to take on a particular value. Unlike the PMF, the value of the PDF at a specific point is not the probability itself but rather a value that, when integrated over an interval, gives the probability that the random variable falls within that interval.</p>
    <p>For a continuous random variable \(X\), the PDF \(f_X(x)\) is defined such that the probability that \(X\) falls within the interval \([a, b]\) is given by:</p>
    <pre>P(a \leq X \leq b) = \int_a^b f_X(x) \, dx</pre>

<h3>Example</h3>
    <p>Consider the standard normal distribution with the PDF:</p>
    <pre>f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}</pre>
    <p>This function describes the likelihood of a standard normal random variable (with mean 0 and variance 1) taking on a particular value.</p>
    <p>For example, the probability that \(X\) falls between -1 and 1 is:</p>
    <pre>P(-1 \leq X \leq 1) = \int_{-1}^{1} \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} \, dx</pre>
    <p>This integral can be evaluated using numerical methods to find the probability.</p>

<h4>Conclusion</h4>
    <p>In summary, the PMF is used for discrete random variables to give the exact probability of specific outcomes, while the PDF is used for continuous random variables to describe the likelihood of outcomes over an interval.</p>

<h4 style='color: blue;'>Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?</h4>
<p><b>Answer:</b></p>
 <h3>What is Cumulative Distribution Function (CDF)?</h3>
    <p>The cumulative distribution function (CDF) of a random variable <i>X</i> is a function that gives the probability that <i>X</i> takes on a value less than or equal to <i>x</i>. Mathematically, the CDF is defined as:</p>
    <p><b>F<sub>X</sub>(x) = P(X ≤ x)</b></p>
    <p>where <b>F<sub>X</sub>(x)</b> is the CDF of the random variable <i>X</i> evaluated at <i>x</i>, and <b>P(X ≤ x)</b> represents the probability that <i>X</i> takes on a value less than or equal to <i>x</i>.</p>
    
<h3>Example</h3>
    <p>Let's consider an example to understand the CDF better.</p>
    <h4>Example for a Discrete Random Variable:</h4>
    <p>Suppose <i>X</i> is a discrete random variable representing the outcome of a fair six-sided die. The probability mass function (PMF) is <b>P(X = x) = 1/6</b> for <i>x</i> in {1, 2, 3, 4, 5, 6}. The CDF <b>F<sub>X</sub>(x)</b> is:</p>
    <p>
        <b>F<sub>X</sub>(x) =</b><br>
        0 if x < 1<br>
        1/6 if 1 ≤ x < 2<br>
        2/6 if 2 ≤ x < 3<br>
        3/6 if 3 ≤ x < 4<br>
        4/6 if 4 ≤ x < 5<br>
        5/6 if 5 ≤ x < 6<br>
        1 if x ≥ 6
    </p>


<h3>Why is CDF Used?</h3>
    <p>The CDF is a fundamental tool in probability theory and statistics because it provides a complete description of the distribution of a random variable. Here are some reasons why CDF is used:</p>
    <ul>
        <li><b>Probability Calculation:</b> It helps in finding the probability that a random variable will take on a value within a specified range.</li>
        <li><b>Distribution Analysis:</b> It provides insight into the distribution and spread of the data.</li>
        <li><b>Quantile Determination:</b> It is used to determine quantiles, which are values that partition the data into equal-sized intervals.</li>
        <li><b>Comparison of Distributions:</b> It allows for the comparison of different distributions by analyzing their CDFs.</li>
    </ul>
    
<p>The CDF is a useful and versatile function in both theoretical and applied statistics, providing a comprehensive way to describe and analyze the behavior of random variables.</p>

<h3 style='color: blue;'>Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.</h3>
<p><b>Answer:</b></p>
   <ul>
        <li><strong>Height of People:</strong> The heights of adult men and women are often modeled using a normal distribution. When collected in large samples, they tend to form a bell-shaped curve centered around an average height.</li>
        <li><strong>IQ Scores:</strong> Intelligence quotient (IQ) scores are designed to follow a normal distribution with a mean of 100 and a standard deviation of 15. This ensures that most people score around the average, with fewer individuals achieving very high or very low scores.</li>
        <li><strong>Measurement Errors:</strong> In various scientific experiments and engineering processes, the errors or deviations from true measurements often follow a normal distribution because the errors tend to accumulate from many small, independent sources.</li>
        <li><strong>Stock Market Returns:</strong> Daily stock returns for a given stock or index are often assumed to be normally distributed. Although this is a simplification, it serves as a useful approximation for many purposes.</li>
        <li><strong>Blood Pressure:</strong> The distribution of systolic blood pressure readings in a large population can be modeled using a normal distribution, with most readings clustering around the mean and fewer readings appearing at the extreme high and low ends.</li>
    </ul>
    
 <h3>Parameters of the Normal Distribution</h3>
    <ul>
        <li><strong>Mean (\(\mu\)):</strong> The mean is the central value around which the data is distributed. It determines the location of the peak of the bell curve. A higher mean shifts the entire distribution to the right, while a lower mean shifts it to the left.</li>
        <li><strong>Standard Deviation (\(\sigma\)):</strong> The standard deviation measures the spread or dispersion of the distribution. It determines the width of the bell curve. A larger standard deviation results in a wider and flatter curve, indicating that the data points are spread out over a larger range of values. A smaller standard deviation results in a narrower and taller curve, indicating that the data points are closely clustered around the mean.</li>
    </ul>
    
<h3>Shape of the Normal Distribution</h3>
    <ul>
        <li><strong>Symmetry:</strong> The normal distribution is perfectly symmetric about its mean. This means that the left half of the distribution is a mirror image of the right half.</li>
        <li><strong>Bell Curve:</strong> The shape of the normal distribution is often referred to as a bell curve due to its bell-like appearance. The curve starts low, rises to a peak at the mean, and then falls symmetrically back down.</li>
        <li><strong>Tails:</strong> The tails of the normal distribution extend infinitely in both directions. However, the probability of extreme values far from the mean decreases rapidly.</li>
    </ul>

<h3 style='color: blue;'>Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?</h3>
<p><b>Answer:</b></p>
 <h3>Bernoulli Distribution</h3>
    <p>The Bernoulli distribution is a discrete probability distribution for a random variable which takes the value 1 with probability \(p\) and the value 0 with probability \(1 - p\). It models a single experiment with two possible outcomes: success (1) and failure (0).</p>
    
<h3>Example</h3>
    <p>Consider a single coin flip where:</p>
    <ul>
        <li>Probability of heads (success, 1): \(p = 0.5\)</li>
        <li>Probability of tails (failure, 0): \(1 - p = 0.5\)</li>
    </ul>
    <p>The outcome of this coin flip follows a Bernoulli distribution with \(p = 0.5\).</p>
 
 <h3>Difference Between Bernoulli Distribution and Binomial Distribution</h3>
    <ul>
        <li><strong>Bernoulli Distribution:</strong>
            <ul>
                <li>Models a single trial with two possible outcomes.</li>
                <li>Has two parameters: \(p\) (probability of success) and \(1 - p\) (probability of failure).</li>
                <li>Example: Flipping a coin once.</li>
            </ul>
        </li>
        <li><strong>Binomial Distribution:</strong>
            <ul>
                <li>Models the number of successes in a fixed number of independent Bernoulli trials.</li>
                <li>Has two parameters: \(n\) (number of trials) and \(p\) (probability of success in each trial).</li>
                <li>Example: Flipping a coin 10 times and counting the number of heads.</li>
            </ul>
        </li>
    </ul>

<h3 style='color: blue;'>Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
is normally distributed, what is the probability that a randomly selected observation will be greater
than 60? Use the appropriate formula and show your calculations.</h3>
<p><b>Answer:</b></p>
   <p>Consider a dataset with a mean(mu) of 50 and a standard deviation (sigma) of 10. We want to find the probability that a randomly selected observation will be greater than 60.</p>
    
   <h4>Steps to Calculate the Probability</h4>
    <ol>
        <li>First, we need to calculate the z-score for 60. The z-score formula is:
            <p>
            z = frac{X - mu}{sigma}
            </p>
            where (X) is the value of interest (60), (mu) is the mean (50), and (sigma) is the standard deviation (10).</li>
        <li>Substitute the given values into the formula:
            <p>
            z = frac{60 - 50}{10} = frac{10}{10} = 1
        </p>
        </li>
        <li>Next, we need to find the probability that a z-score is greater than 1. This can be found using standard normal distribution tables or a calculator.</li>
        <li>Using the standard normal distribution table, we find the cumulative probability for (z = 1):
            <p>
            P(Z leq 1) = 0.8413
            </p>
        </li>
        <li>To find the probability that (Z) is greater than 1, we subtract the cumulative probability from 1:
            <p>
            P(Z > 1) = 1 - P(Z leq 1) = 1 - 0.8413 = 0.1587
            </p>
        </li>
    </ol>
    
<h4>Result</h4>
    <p>The probability that a randomly selected observation from this normally distributed dataset will be greater than 60 is 0.1587, or 15.87%.</p>


<h3 style='color: blue;'>Q7: Explain uniform Distribution with an example.</h3>
<p><b>Answer:</b></p>
   <h3>Uniform Distribution</h3>
    <p>The uniform distribution is a type of probability distribution in which all outcomes are equally likely. Every value within a specified range has an equal probability of occurring, and the probability is zero for values outside this range.</p>

<h4>Example</h4>
    <p>Consider the roll of a fair six-sided die. The die has six faces, numbered from 1 to 6. Since the die is fair, each face has an equal probability of landing face up. This situation can be modeled using a uniform distribution.</p>

<ul>
        <li><strong>Range of Values:</strong> The possible outcomes are the integers 1, 2, 3, 4, 5, and 6.</li>
        <li><strong>Equal Probability:</strong> Each outcome has a probability of (frac{1}{6}).</li>
    </ul>

<h4>Probability Density Function (PDF)</h4>
    <p>For a continuous uniform distribution defined on the interval ([a, b]), the probability density function (PDF) is:</p>
    <pre>
    f(x) = {
        1 / (b - a)   if a ≤ x ≤ b
        0             otherwise
    }
    </pre>

<h4>Discrete Uniform Distribution</h4>
    <p>For the example of the six-sided die, which is a discrete uniform distribution, the probability mass function (PMF) is:</p>
    <pre>
    P(X = x) = {
        1 / 6   if x ∈ {1, 2, 3, 4, 5, 6}
        0       otherwise
    }
    </pre>

<h4>Visual Representation</h4>
    <p>If we plot the probabilities of each outcome for the roll of a fair six-sided die, we get a bar graph with six bars of equal height:</p>
    <ul>
        <li>Each bar represents one of the outcomes (1, 2, 3, 4, 5, 6).</li>
        <li>The height of each bar is (frac{1}{6}), indicating the equal probability of each outcome.</li>
    </ul>



<h3 style='color: blue;'>Q8: What is the z score? State the importance of the z score.</h3>
<p><b>Answer:</b></p>
<h4>What is the Z Score?</h4>
    <p>
        The z score, also known as the standard score, is a statistical measurement that describes a value's position relative to the mean of a group of values. It is calculated by taking the difference between the value and the mean, and then dividing by the standard deviation. The formula for the z score is:
    </p>
    <p>
        <code>z = (X - μ) / σ</code>
    </p>
    <p>
        where:
        <ul>
            <li><strong>X</strong> is the value being measured.</li>
            <li><strong>μ</strong> is the mean of the population.</li>
            <li><strong>σ</strong> is the standard deviation of the population.</li>
        </ul>
    </p>

<h4>Importance of the Z Score</h4>
    <ul>
        <li><strong>Standardization:</strong> The z score allows for standardization of different data points from different normal distributions, making it possible to compare scores across different scales.</li>
        <li><strong>Identifying Outliers:</strong> Z scores help identify outliers in data. Typically, a z score above 3 or below -3 indicates an outlier.</li>
        <li><strong>Probability Calculation:</strong> Z scores are used to calculate probabilities in a standard normal distribution. This is useful in hypothesis testing and confidence interval estimation.</li>
        <li><strong>Interpreting Scores:</strong> Z scores provide a way to interpret the relative position of a data point within a distribution. A positive z score indicates a value above the mean, while a negative z score indicates a value below the mean.</li>
        <li><strong>Transforming Data:</strong> Z scores can be used to transform data to a standard normal distribution, which is a key step in various statistical analyses and methods, such as in regression analysis and control charts.</li>
    </ul>

<h3 style='color: blue;'>Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.</h3>
<p><b>Answer:</b></p>
 <h4>Central Limit Theorem (CLT)</h4>
    <p>The Central Limit Theorem (CLT) is a fundamental theorem in probability theory and statistics. It states that when independent random variables are added together, their normalized sum tends toward a normal distribution (also known as a Gaussian distribution), regardless of the original distribution of the variables themselves.</p>
    
<h4>Significance of the Central Limit Theorem</h4>
    <p>The significance of the Central Limit Theorem lies in its broad applicability and several key implications:</p>
 <ul>
        <li><strong>Approximation of Distributions:</strong> It allows us to approximate the distribution of sample means (or sums) from any distribution with a normal distribution, especially useful when the original distribution is unknown or complex.</li>
        <li><strong>Inference:</strong> It forms the basis for many statistical techniques, such as hypothesis testing and confidence interval construction, because it provides a way to understand the distribution of sample statistics.</li>
        <li><strong>Real-world Applications:</strong> CLT is applied extensively in fields such as quality control, finance, engineering, and social sciences where understanding the behavior of sample means or sums is crucial.</li>
        <li><strong>Robustness:</strong> Even if the individual variables do not follow a normal distribution, as long as they are independent and identically distributed (i.i.d.), their sum will approach a normal distribution with sufficient sample size.</li>
    </ul>
    
<p>Overall, the Central Limit Theorem is a cornerstone of statistical theory, providing a bridge between the characteristics of individual observations and the properties of their sums or averages in large samples.</p>


<h3 style='color: blue;'>Q10: State the assumptions of the Central Limit Theorem.</h3>
<p><b>Answer:</b></p>
 <h4>Assumptions of the Central Limit Theorem</h4>
    <ul>
        <li><strong>Independence:</strong> The observations or data points should be independent of each other. This means that the occurrence of one observation does not affect the occurrence of another.</li>
        <li><strong>Sample Size:</strong> The sample size should be sufficiently large. There is no strict rule on the minimum sample size, but a larger sample size tends to yield better approximations.</li>
        <li><strong>Identical Distribution:</strong> The observations should come from the same probability distribution. This distribution can be any distribution, not necessarily normal, but the CLT tells us the distribution of the sample mean approaches normality as the sample size increases.</li>
        <li><strong>Finite Variance:</strong> The population should have a finite variance. This assumption ensures that the mean of the sample means converges to the population mean as the sample size increases.</li>
    </ul>
