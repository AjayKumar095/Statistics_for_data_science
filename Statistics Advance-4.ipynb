<h3 style="color: blue;">Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.</h3>
<p><b>Answer:</b></p>


<p>Bayes' Theorem is a fundamental concept in probability theory that allows us to update the probability of a hypothesis (event) given new evidence. The formula is:</p>

<p style="margin-left: 20px;"><strong>P(A|B) = (P(B|A) * P(A)) / P(B)</strong></p>

<p>Where:</p>
<ul>
    <li> P(A|B): Posterior probability of event A given evidence B.</li>
    <li> P(B|A): Probability of evidence B given event A (likelihood).</li>
    <li> P(A): Prior probability of event A.</li>
    <li> P(B): Probability of evidence B (normalizing constant).</li>
</ul>

<p>Example:</p>

<p>A factory produces two types of products: Product A and Product B. Product A has a defect rate of 5%, and Product B has a defect rate of 2%. Given a randomly selected defective product, what is the probability that it is Product A?</p>

<p>We want to find P(A|D), the probability of Product A given the defect.</p>

<p>Let:</p>
<ul>
    <li> P(A): Probability of selecting Product A = 0.5 </li>
    <li> P(B): Probability of selecting Product B = 0.5 </li>
    <li> P(D|A): Probability of defect given Product A = 0.05 </li>
    <li> P(D|B): Probability of defect given Product B = 0.02 </li>
</ul>

<p>First, calculate P(D), the probability of the defect:</p>

<p>P(D) = P(D|A) * P(A) + P(D|B) * P(B)</p>

<p>P(D) = 0.05 * 0.5 + 0.02 * 0.5</p>

<p>P(D) = 0.025 + 0.01</p>

<p>P(D) = 0.035</p>

<p>Now calculate P(A|D), the probability that a randomly selected defective product is Product A:</p>

<p>P(A|D) = (P(D|A) * P(A)) / P(D)</p>

<p>P(A|D) = (0.05 * 0.5) / 0.035</p>

<p>P(A|D) = 0.025 / 0.035</p>

<p>P(A|D) ‚âà 0.714 or 71.4%</p>

<p>Therefore, the probability that a randomly selected defective product is Product A is approximately 71.4%. This calculation demonstrates how Bayes' Theorem allows us to update our prior belief (prior probability) about which product is defective given new evidence (likelihood of the defect).</p>
