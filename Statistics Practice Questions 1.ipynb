```python
import numpy as np 
import math
import warnings
warnings.filterwarnings('ignore')
## class that is used to find the mean, median, modes, trimmed mean, midrange mean, harmonic mean, geometric mean and weighted mean.
class statistic:
    
    def __init__(self) -> None:
        pass
    
    def Mean(self, num_list:list):
        Sum=0
        for num in num_list:
            Sum += num
        
        mean=Sum/len(num_list)  
        return mean
    
    def Median(self, num_list:list):
        lenght=len(num_list)
        mid=lenght//2
        num_list=np.sort(num_list)
        if len(num_list)%2==0:
            median=(num_list[mid-1]+num_list[mid])/2
            return median
        
        else:
            median=num_list[mid]
            return median
    
    def Mode(self, num_list:list):
            frequency = {}
            for number in num_list:
                 if number in frequency:
                     frequency[number] += 1
                 else:
                     frequency[number] = 1

            max_count = max(frequency.values())
            modes = [key for key, value in frequency.items() if value == max_count]

            if len(modes) == len(num_list):
                 return "No mode found, each value appears only once."
            return modes
    
    def weighted_mean(self, num_list:list, weights:list):
        if len(num_list) != len(weights):
            return 'Lenght does not match.'
        
        weight_sum=sum(value*weight for value, weight in zip(num_list, weights))  
        total_weight=sum(weights)
        
        if total_weight==0:
            return ValueError('Weights sum not be zero.')
        
        return weight_sum/total_weight
    
    def geometric_mean(self, num_list:list):
          if any(value == 0 for value in num_list):
            raise ValueError("The list contains zero, which will result in a geometric mean of zero.")
    
          log_sum = sum(math.log(value) for value in num_list)
          n = len(num_list)

          return math.exp(log_sum / n) 
    
    def harmonic_mean(self, num_list:list):
            lenght = len(num_list)
            reciprocal_sum = sum(1 / num for num in num_list)
    
            return lenght / reciprocal_sum 
    
    def midrange_mean(self, num_list:list):
        
        if len(num_list) == 0:
            return ValueError('Empty List.')
        
        min_value=min(num_list)
        max_value=max(num_list)
        
        return (min_value+max_value)/2
    
    def trimmed_mean(self, num_list:list, percentage:int):
        if not 0< percentage < 50:
            return ValueError('percentage must be between 0 to 50.')
        
        if len(num_list) == 0:
            return ValueError('Empty List.')
        
        num_list=np.sort(num_list)
        lenght=len(num_list)
        trim_count = int(lenght * (percentage / 100))

        trimmed_numbers = num_list[trim_count:lenght - trim_count]
        return sum(trimmed_numbers) / len(trimmed_numbers)
    
## genertaing a list of 100 integers containing values between 90 to 130.
int_list = np.random.randint(90, 130, 100)  
weights=np.random.randint(1,40, 100)



print('Mean of the given integer number list:', statistic().Mean(num_list=int_list))
print('Median of the given integer number list:', statistic().Median(num_list=int_list))
print('Mode of the given integer number list:', statistic().Mode(num_list=int_list))
print('Weighted mean of the given integer number list:', statistic().weighted_mean(num_list=int_list, weights=weights))
print('Geometric mean of the given integer number list:', statistic().geometric_mean(num_list=int_list))
print('Harmonic mean of the given integer number list:', statistic().harmonic_mean(num_list=int_list))
print('Trimmed mean of the given integer number list:', statistic().trimmed_mean(num_list=int_list, percentage=20))
           
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import norm, gaussian_kde

# Step 1: Generate a list of 500 integers between 200 and 300
int_list2 = np.random.randint(200, 301, 500)

# Step 2: Calculate mean and standard deviation of the data
mean = np.mean(int_list2)
std_dev = np.std(int_list2)

# Step 3: Generate points for the Gaussian (normal) distribution curve
x_gaussian = np.linspace(np.min(int_list2), np.max(int_list2), 100)
gaussian_curve = norm.pdf(x_gaussian, mean, std_dev)

kde = gaussian_kde(int_list2)
x_kde = np.linspace(np.min(int_list2), np.max(int_list2), 100)
kde_curve = kde.evaluate(x_kde)

fig, ax = plt.subplots(1,3,figsize=(30,10))
sns.histplot(int_list2, ax=ax[0])
ax[0].set_title('Frequency')

sns.lineplot(x=x_gaussian, y=gaussian_curve, ax=ax[1])
ax[1].set_title('gaussian')

sns.lineplot(x=x_kde, y=kde_curve, ax=ax[2])
ax[2].set_title('KED')


```

```python

class descriptive_stat:
    
    def __init__(self) -> None:
        pass
    
    def Cal_Range(self, num_list:list):
        if len(num_list)==0:
            return 'Empty list'
        return  max(num_list) - min(num_list) 
    
    def variance_and_std(self, num_list:list):
        if len(num_list)==0:
            return 'Empty list'
        
        length=len(num_list)
        mean=np.mean(num_list)
        Sum=0
        for num in num_list:
            Sum += (num-mean)**2
        
        variance=Sum/length
        
        std=np.sqrt(variance)
        return (variance, std)
            
            
        
var, std= descriptive_stat().variance_and_std([1,2,3,4,5])

print(f'Variance of the given int list:{var}\nStd of give int list:{std}')
```
